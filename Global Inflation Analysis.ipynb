```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r"E:\Data Projects\Global Inflation Analysis\global_inflation_data.csv")
```

```python
df = data.copy()
```

```python
df.describe(include='all')
```

```python
df['country_name'] = df['country_name'].astype('category')
df['indicator_name'] = df['indicator_name'].astype('category')
```

```python
df.info()
```

```python
df
```

```python
df.columns
```

```python
#missing_1980 = df_2['1980'].mean() 
#instead of one by one i use loop 

# Fill missing values year by year using their own mean
for year in range(1980, 2025): 
    col = str(year)
    if col in df.columns:
        df[col] = df[col].fillna(df[col].mean())
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.head()
```

```python
# melt is used convert wide format in to long format which make easy to analyse
df = pd.melt(
    df,
    id_vars=['country_name', 'indicator_name'],  # Columns to keep as they are
    var_name='year',                             # Name of the new column holding years
    value_name='inflation_rate'                  # Name of the new column holding values
)

# Convert year to numeric for analysis
df['year'] = pd.to_numeric(df['year'], errors='coerce')
```

```python
df
```

```python
# Filter only inflation indicator rows
df_inflation = df[df['indicator_name'].str.contains('inflation', case=False)]

#Filter for Pakistan only
df_pakistan = df_inflation[df_inflation['country_name'] == 'Pakistan']
```

```python
df_pakistan.head()
```

```python
print(df_pakistan.isnull().sum())
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=df_pakistan, x='year', y='inflation_rate')
plt.title('Pakistan Inflation Rate Over Time')
plt.ylabel('Inflation Rate (%)')
plt.xlabel('Year')
plt.show()
```

```python
global_inflation = df.groupby('year')['inflation_rate'].mean().reset_index()
print(global_inflation.head())

plt.figure(figsize=(12,6))
sns.lineplot(data=global_inflation, x='year', y='inflation_rate')
plt.title('Global Average Inflation Rate Over Time')
plt.ylabel('Average Inflation Rate (%)')
plt.xlabel('Year')
plt.show()
```

```python
global_inflation['decade'] = (global_inflation['year'] // 10) * 10

# Calculate average inflation per decade
decade_avg = global_inflation.groupby('decade')['inflation_rate'].mean().reset_index()

plt.figure(figsize=(12,6))
sns.barplot(data=decade_avg, x='decade', y='inflation_rate', palette='viridis')
plt.title('Global Average Inflation Rate per Decade')
plt.xlabel('Decade')
plt.ylabel('Average Inflation Rate (%)')
plt.show()
```

```python
x = global_inflation['year']
y = global_inflation['inflation_rate']

plt.figure(figsize=(12,6))

# Use seaborn style
sns.set(style="whitegrid")

# Scatter plot using matplotlib
plt.scatter(x, y, color='blue', label='Average Inflation')

# Fit and plot trend line
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x, p(x), "r--", label='Trend Line')

plt.title('Global Average Inflation Rate Over Time')
plt.xlabel('Year')
plt.ylabel('Average Inflation Rate (%)')
plt.legend()
plt.show()
```
